# CIFAR-10 Classification

## Imports

```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import numpy as np
import sklearn as sk
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
from sklearn import svm
from sklearn import linear_model
from sklearn.neural_network import MLPClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, f1_score
```

## Load Data

```python
def load_cifar10_data():
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
    print("Total Shape of Train Data:", np.shape(x_train))
    print("Total Shape of Train Label:", np.shape(y_train))
    print("Total Shape of Test Data:", np.shape(x_test))
    print("Total Shape of Test Label:", np.shape(y_test))
    return x_train, y_train, x_test, y_test
```

## Representations

```python
class Representation:
    def __init__(self, x_train, y_train, x_test, y_test, train_num=200, test_num=200):
        x_train = tf.image.rgb_to_grayscale(x_train)
        self.x_train_all = tf.divide(tf.to_float(x_train), tf.constant(255.0)).eval(session=tf.Session())
        self.y_train_all = y_train
        x_test = tf.image.rgb_to_grayscale(x_test)
        self.x_test_all = tf.divide(tf.to_float(x_test), tf.constant(255.0)).eval(session=tf.Session())
        self.y_test_all = y_test
        self.train_shape_all = np.shape(self.x_train_all)
        self.test_shape_all = np.shape(self.x_test_all)
        self.train_shape = np.array([train_num, self.train_shape_all[1], self.train_shape_all[2], self.train_shape_all[3]])
        self.test_shape = np.array([test_num, self.test_shape_all[1], self.test_shape_all[2], self.test_shape_all[3]])
        self.x_train = self.y_train = self.x_test = self.y_test = None
        self.train_rows = self.test_rows = None
        print("Transformed Train Data:", np.shape(self.x_train))
        print("Transformed Train Label:", np.shape(self.y_train))
        print("Transformed Test Data:", np.shape(self.x_test))
        print("Transformed Test Label:", np.shape(self.y_test))

    def get_random(self):
        self.train_rows = np.random.randint(0, self.train_shape_all[0], size=self.train_shape[0]).astype(np.int32)
        self.test_rows = np.random.randint(0, self.test_shape_all[0], size=self.test_shape[0]).astype(np.int32)
        self.x_train = self.x_train_all[self.train_rows]
        self.y_train = self.y_train_all[self.train_rows]
        self.x_test = self.x_test_all[self.test_rows]
        self.y_test = self.y_test_all[self.test_rows]
        
    def get_raw(self):
        self.get_random()
        return (self.x_train, self.y_train, self.x_test, self.y_test)
    
    def get_flatten(self):
        self.get_random()
        x_train_flat = tf.reshape(self.x_train, [self.train_shape[0], self.train_shape[1]*self.train_shape[2]*self.train_shape[3]]).eval(session=tf.Session())
        x_test_flat = tf.reshape(self.x_test, [self.test_shape[0], self.test_shape[1]*self.test_shape[2]*self.test_shape[3]]).eval(session=tf.Session())
        return (x_train_flat, self.y_train, x_test_flat, self.y_test)

    def get_pca(self, num_components):
        self.get_random()
        x_train_pca = tf.reshape(self.x_train, [self.train_shape[0], self.train_shape[1]*self.train_shape[2]*self.train_shape[3]]).eval(session=tf.Session())
        x_test_pca = tf.reshape(self.x_test, [self.test_shape[0], self.test_shape[1]*self.test_shape[2]*self.test_shape[3]]).eval(session=tf.Session())
        pca = PCA(n_components=num_components)
        x_pca = pca.fit(x_train_pca)
        x_train_new = x_pca.transform(x_train_pca)
        x_test_new = x_pca.transform(x_test_pca)
        return (x_train_new, self.y_train, x_test_new, self.y_test)

    def get_lda(self, num_components):
        self.get_random()
        x_train_lda = tf.reshape(self.x_train, [self.train_shape[0], self.train_shape[1]*self.train_shape[2]*self.train_shape[3]]).eval(session=tf.Session())
        x_test_lda = tf.reshape(self.x_test, [self.test_shape[0], self.test_shape[1]*self.test_shape[2]*self.test_shape[3]]).eval(session=tf.Session())
        lda = LinearDiscriminantAnalysis(n_components=num_components)
        x_lda = lda.fit(x_train_lda, self.y_train)
        x_train_new = x_lda.transform(x_train_lda)
        x_test_new = x_lda.transform(x_test_lda)
        return (x_train_new, self.y_train, x_test_new, self.y_test)
    
    def get_lda_model(self):
        x_train_lda = tf.reshape(self.x_train, [self.train_shape[0], self.train_shape[1]*self.train_shape[2]*self.train_shape[3]]).eval(session=tf.Session())
        x_test_lda = tf.reshape(self.x_test, [self.test_shape[0], self.test_shape[1]*self.test_shape[2]*self.test_shape[3]]).eval(session=tf.Session())
        lda = LinearDiscriminantAnalysis()
        x_lda = lda.fit(x_train_lda, self.y_train)
        return lda
    
    def get_tsne(self, num_components):
        self.get_random()
        x_train_tsne = tf.reshape(self.x_train, [self.train_shape[0], self.train_shape[1]*self.train_shape[2]*self.train_shape[3]]).eval(session=tf.Session())
        x_test_tsne = tf.reshape(self.x_test, [self.test_shape[0], self.test_shape[1]*self.test_shape[2]*self.test_shape[3]]).eval(session=tf.Session())
        tsne = TSNE(n_components=num_components, init='pca')
        x_train_new = tsne.fit_transform(x_train_tsne)
        x_test_new = tsne.fit_transform(x_test_tsne)
        return (x_train_new, self.y_train, x_test_new, self.y_test)
    
    
```

## Classifier Models

```python
class Models:
    def __init__(self, x_train, y_train, x_test, y_test):
        self.x_train = x_train
        self.y_train = y_train
        self.x_test  = x_test
        self.y_test  = y_test
    
    def linear_svm(self, penalty='l2', C=1, loss='squared_hinge'):
        clf = svm.LinearSVC(penalty=penalty, C=C)
        clf.fit(self.x_train, self.y_train)
        pred = np.rint(clf.predict(self.x_test))
        return (f1_score(self.y_test, pred, average='macro'), accuracy_score(self.y_test, pred))

    
    def kern_svm(self, kern='rbf', C=1):
        clf = svm.SVC(kernel=kern, C=C)
        clf.fit(self.x_train, self.y_train)
        pred = np.rint(clf.predict(self.x_test))
        return (f1_score(self.y_test, pred, average='macro'), accuracy_score(self.y_test, pred))

    def linear_classifier(self, norm='l2', solver='sag', reg=1):
        clf = linear_model.LogisticRegression(penalty=norm, C=reg, solver=solver)
        clf.fit(self.x_train, self.y_train)
        pred = np.rint(clf.predict(self.x_test))
        return (f1_score(self.y_test, pred, average='macro'), accuracy_score(self.y_test, pred))

    def mlp(self, solver='adam', activation='relu', h_size=(100,), eta=0.001, l_mode='constant', b_1=0.9, b_2=0.999, eps=1e-8):
        clf = MLPClassifier(solver=solver, activation=activation, hidden_layer_sizes=h_size, learning_rate_init=eta,
                            learning_rate=l_mode, beta_1=b_1, beta_2=b_2, epsilon=eps)
        clf.fit(self.x_train, self.y_train)
        pred = np.rint(clf.predict(self.x_test))
        return (f1_score(self.y_test, pred, average='macro'), accuracy_score(self.y_test, pred))

    def get_dt(self, min_impurity_dec):
        clf = tree.DecisionTreeClassifier(min_impurity_decrease=min_impurity_dec)
        clf.fit(self.x_train, self.y_train)
        pred = np.rint(clf.predict(self.x_test))
        return (f1_score(self.y_test, pred, average='macro'), accuracy_score(self.y_test, pred))
```

## Data Loads

```python
train_num = 2500
test_num = 800
x_train_raw, y_train_raw, x_test_raw, y_test_raw = load_cifar10_data()
data = Representation(x_train_raw, y_train_raw, x_test_raw, y_test_raw, train_num=train_num, test_num=test_num)
```

## Hyperparameter Selection


### Linear SVM

```python
# Creating model for RAW DATA
X, Y, X_t, Y_t = data.get_flatten()
model_raw = Models(X, Y, X_t, Y_t)

c_vals = np.linspace(0.001, 1.0, 100)
avg_acc = []
f_score = []
for c in c_vals:
    a, b = model_raw.linear_svm(C=c)
    avg_acc.append(b)
    f_score.append(a)
```

#### Related Graphs and Analysis

Naive method of finding the penalty coeffiecient is to try out each valu of `C` at specific small intervals
but this takes in too much time and hence not feasible. So, to improve it we would like to first search for
values at broader gaps and then try to decrease the gap distance bit-by-bit and finally arrive at the answer.
Assumption here is that the graph of Average accuracy vs `C` will be a convex curve which is fairly reasonable
to assume in case of SVMs(Large `C` means overfitting will take place and small will decrease the accuracy).
We need to select `C` which generalizes well.

**NOTE** : *I have used only 1500 random samples for training and 500 random samples for testing, so accuracy
and f1-score is quite low.*

```python
plt.figure()
plt.plot(c_vals, avg_acc)
plt.title('Avg Acc vs C')
plt.ylabel('Avg. Accuracy')
plt.xlabel('C')
plt.figure()
plt.plot(c_vals, f_score)
plt.title('F1 Score vs C')
plt.ylabel('F Score')
plt.xlabel('C')
```

As we can see that best F1-Score and Average accuracy occurs around `C = 0.05` with window of around `0.01`.
We could use this information in better calculating the value for best `C` value for PCA and LDA based
dimensionally reduced SVM classification. Also to select the number of components to be selected can be
derived by calculating reconstruction error. If reconstruction error is < 10% then it would be a good choice.
In my case I have selected 600 features which gave around 5.23 % error of reconstruction.

```python
# Creating model for PCA DATA
X, Y, X_t, Y_t = data.get_pca(num_components=600)
model_raw = Models(X, Y, X_t, Y_t)

c_vals = np.linspace(0.00001, 0.18, 250)
avg_acc = []
f_score = []
for c in c_vals:
    a, b = model_raw.linear_svm(C=c)
    avg_acc.append(b)
    f_score.append(a)
```

#### Related Graphs and Analysis

```python
plt.figure()
plt.plot(c_vals, avg_acc)
plt.title('Avg Acc vs C')
plt.ylabel('Avg. Accuracy')
plt.xlabel('C')
plt.figure()
plt.plot(c_vals, f_score)
plt.title('F1 Score vs C')
plt.ylabel('F Score')
plt.xlabel('C')
```

Required value for `C` is between `0.010` to `0.015`. I am going to carry this over when I use Linear SVM on LDA based representation. Ideally, we should try to do this the reverse way (starting from lower dimensional representation
and going to higher dimension representation) to save more time. **Although this should not work because LDA is
supervised dimensionality reduction.**

```python
# Creating model for PCA DATA
X, Y, X_t, Y_t = data.get_lda(num_components=9)
model_raw = Models(X, Y, X_t, Y_t)

c_vals = np.linspace(0.00001, 1, 450)
avg_acc = []
f_score = []
for c in c_vals:
    a, b = model_raw.linear_svm(C=c)
    avg_acc.append(b)
    f_score.append(a)
```

#### Related Graphs and Analysis

```python
plt.figure()
plt.plot(c_vals, avg_acc)
plt.title('Avg Acc vs C')
plt.ylabel('Avg. Accuracy')
plt.xlabel('C')
plt.figure()
plt.plot(c_vals, f_score)
plt.title('F1 Score vs C')
plt.ylabel('F Score')
plt.xlabel('C')
```

So, in fact my assumption was correct. The chosen `C` value for pca or raw data will not work. We need to penalize
the wrong classifications more to get a better model because LDA already tries to separate the data upon changing the representation. Also we need to find the optimal number of components to be used for dimensionality reduction. So to
select the number of dimensions to be used I would like to keep my variance above 90%.(Here this part is insignificant because **LDA** already reduces the dimension to `num_class - 1`. But similar procedure was used for PCA.)

```python
num_comp = 0
var_sum = 0

lda = data.get_lda_model()

while var_sum < 0.95:
    var_sum += lda.explained_variance_ratio_[num_comp]
    num_comp += 1

print("num_comp:", num_comp)
```

```python
X, Y, X_t, Y_t = data.get_lda(num_components=num_comp)
model_raw = Models(X, Y, X_t, Y_t)

c_vals = np.linspace(0.1, 10, 500)
avg_acc = []
f_score = []
for c in c_vals:
    a, b = model_raw.linear_svm(C=c)
    avg_acc.append(b)
    f_score.append(a)
```

```python
plt.figure()
plt.plot(c_vals, avg_acc)
plt.title('Avg Acc vs C')
plt.ylabel('Avg. Accuracy')
plt.xlabel('C')
plt.figure()
plt.plot(c_vals, f_score)
plt.title('F1 Score vs C')
plt.ylabel('F Score')
plt.xlabel('C')
```

Clearly the `C` of SVM should be less than between 2 to 3 . Also we do not get that much accuracy using LDA because data is not separable using current features. We will need to use KLDA or KPCA for better class separation. Although, if we see the performance comparing the number of features we are using ve average accuracy LDA is quite good.

```python
c_vals = np.linspace(0.0002, 0.6, 500)
avg_acc = []
f_score = []
for c in c_vals:
    a, b = model_raw.linear_svm(C=c)
    avg_acc.append(b)
    f_score.append(a)

def plot_graphs(c_vals, avg_acc, f_score):
    plt.figure()
    plt.plot(c_vals, avg_acc)
    plt.title('Avg Acc vs C')
    plt.ylabel('Avg. Accuracy')
    plt.xlabel('C')
    plt.figure()
    plt.plot(c_vals, f_score)
    plt.title('F1 Score vs C')
    plt.ylabel('F Score')
    plt.xlabel('C')

plot_graphs(c_vals, avg_acc, f_score)
```

As we can see from the graphs that high accuracy and f1 score are achieved around `C=0.2`.

As we can see from the graphs that very low values of `min_impurity_decrease` causes decrease in accuracy and f1-score over testing data. Optimal value of `min_impurity_decrease` is somewhere around `0.005`.

## Problems Faced while Building Classifier from Data

 - Many times some values are missing from columns in the dataset. We have two options: Data Imputation or Data Deletion. Both of them have serious issues. Data Imputation can cause biasing cause we are generating data from other parameters and data deletion can reduce the number of samples we have.
 - Which classifier is best for the data? Some classifier will perform better on a given dataset than on the others. Selection of the classifier is a major task. Generally, we implement multiple classifier models and pick the best one out or combine the results of those classifiers to get another better classifier.
 - For SVMs we need to select the penalty value for each misclassified sample. Higher penalty means that boundary is very constricted and lower penalty leads to bad classifier for genral cases. There is a fine range of values which we need to select to get the best out of SVMs. In, Linear Classifier we have problems such as ending up at local minima and regularisation factor to use. MLPs have problems like - number of layers to use, activation function for each neuron or layer, initial state of the neurons, etc. For decision tree we need to decide when to stop the separation between classes so as to avoid overfitting.
